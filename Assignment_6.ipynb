### Q1. What are escape characters, and how do you use them?

##### Escape characters are represented with a back slash(/) symbol. They are used to escape a special character inside a string. We can use them by adding a back slash followed by a special character inside a string.

### Q2. What do the escape characters n and t stand for?

##### The escape character n stands for new line and t stands for tab.

### Q3. What is the way to include backslash characters in a string?

##### To include a backslash character in a string, you need to add a back slash character extra before it. Which escapes the special charcater. Example is typying "\\t" will returns a back slash character inside a string.

### Q4. The string "Howl's Moving Castle" is a correct value. Why isn't the single quote character in the word Howl's not escaped a problem?

##### It is because the string is represented inside double quotes. Therefore, single quote is not escaped.

### Q5. How do you write a string of newlines if you don't want to use the n character?

### Q9. What are the methods for right-justifying, left-justifying, and centering a string?

##### The method for right-justifying is str.rjust() method. For left-justifying is str.ljust() method and for centering a string use str.center() method.

### Q10. What is the best way to remove whitespace characters from the start or end?

##### For removing whitespace characters from start or end, use .strip() inbuilt function. The strip() function removes the whitespaces from both start and end of the string. You can even specify a character inside the strip() so that, that character will be removed from start and end. If that character is not found, the it returns the same string.
