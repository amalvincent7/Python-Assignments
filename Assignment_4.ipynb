### Q17. How do you distinguish between copy.copy() and copy.deepcopy()?

##### A copy.copy() is termed as shallow copy. The changes are applicable when a list contains another list inside. Example a=[1,2,3,[5,6]], consider it as orginal object. If we type a1=copy.copy(a), a new variable is created with reference to orginal variable a. In this case if we change a value inside the list in list, even the orginal list gets changed. Where as in case of deep copy, this change won't affect the orginal list.
