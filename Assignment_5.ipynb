### Q3. What is the most significant distinction between a dictionary and a list?

##### A dictionary is represented inside curly braces and list is represented inside square braces. Dictionary is combination of key value pairs, while list is having just elements in it. List is ordered and dictionary is unordered.

### Q5. If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and 'cat' in spam.keys()?

##### There is no difference between both, because both returns true if "cat" is available as key inside the dictionary named spam.

### Q6. If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and 'cat' in spam.values()?

##### There is a difference. "cat" in spam will return true, if "cat" is available as key in spam and "cat" in spam.values() will return true, only if "cat" is available as value inside the dictionary.
